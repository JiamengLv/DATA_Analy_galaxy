```python
from PIL import Image
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

def kmeans_segmentation(image_path, num_clusters=3):
   """
   使用KMeans对图像进行分割。
   
   参数:
   - image_path: str, 图像文件路径。
   - num_clusters: int, 聚类数量，默认为3。
   
   返回:
   - 原始图像和聚类标签图像的可视化。
   """

   data = Image.open(image_path)
   data = data.convert("RGBA")
   img = np.array(data)
   height, width, _ = img.shape
   

   datas = img.reshape((height * width, 4))
   

   kmeans = KMeans(n_clusters=num_clusters)
   kmeans.fit(datas)
   labels = kmeans.predict(datas)
   

   labels_img = labels.reshape((height, width))
   
   print("Cluster Labels Shape:", labels_img.shape)
   
   plt.figure(figsize=(12, 6))
   plt.subplot(121)
   plt.imshow(data)
   plt.axis('off')
   plt.title('Original Image')
   
   plt.subplot(122)
   plt.imshow(labels_img, cmap='viridis')
   plt.axis('off')
   plt.title('Segmented Image')
   
   plt.show()

image_path = "/home/dell461/ljm/Multi_data/Code/seg/1741004794898.jpg"
kmeans_segmentation(image_path, num_clusters=4)
```
